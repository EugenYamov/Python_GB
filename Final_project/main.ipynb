Видимо, в задании приведена не та функция. Задача имеет бесконечное число решений.

```python
from matplotlib import pyplot as plt
import math
```

```python
def f(x):
    return (-12)*(x**4)*(math.sin(math.cos(x))) - 18*(x**3) + 5*(x**2) + 10*x - 30

x = [x for x in range (-100, 100)]
y = [f(t) for t in x]
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.minorticks_on()
plt.grid(which='minor')
plt.grid(which='major')
plt.plot(x,y)


```
